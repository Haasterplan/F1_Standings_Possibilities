```python
import sys
import time
import numpy as np
import pandas as pd
import pickle as pkl
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

```python
def SimulateRace(initial_possibilities,samples,points_available_discrete_values):
    values=np.random.choice(initial_possibilities,size=samples)
    values_mg,padv_mg=np.meshgrid(values,points_available_discrete_values)
    print((values_mg+padv_mg).shape)
    possibilities=(values_mg+padv_mg).flatten()
    # unique_values,unique_value_counts=np.unique(possibilities,return_counts=True)
    return possibilities

def SimulateSeason(possibilities,samples,races_remaining,points_avaiable_discrete_values):
    possibilities=SimulateRace(possibilities,samples,points_available_discrete_values)
    if races_remaining==0:
        return np.unique(possibilities,return_counts=True)
    else:
        # print('a')
        races_remaining-=1
    return SimulateSeason(possibilities,samples,races_remaining,points_available_discrete_values)
# initial_values=np.array(list(range(5)))
initial_values=np.array([258])
samples=1000
# initial_counts=np.ones([1])
points_available_discrete_values=np.array([0,25,0,34,26,25,15,25,25,6,27,25,25])
races_remaining=8
values,counts=SimulateSeason(initial_values,samples,races_remaining,points_available_discrete_values)
print(values,counts)
```

```python
def Normalize(data):
	max_val=data.max()
	min_val=data.min()
	data=(data-min_val)/(max_val-min_val)
	return data,max_val,min_val

def ReScale(data,max_val,min_val):
	return data*(max_val-min_val)+min_val
```

```python
from scipy.stats import norm,alpha
data=values
data,max_val,min_val=Normalize(values)
# params=alpha.fit(values)
params=alpha.fit(data)
arg = params[:-2]
loc = params[-2]
scale = params[-1]
# plt.hist(data)
# plt.plot(values,counts/counts.sum())
plt.plot(data,alpha.cdf(data,loc=loc,scale=scale,*arg))
# plt.plot(values,norm.pdf(data,loc=loc,scale=scale))
```

```python
def MC_Step(initial,n):
    initial=(initial+np.random.rand(initial.size))/2
    # initial/=initial.sum()
    if n==0:
        return initial
    else:
        n-=1
    return MC_Step(initial,n)
```

```python
initial=np.random.rand(100)
initial/=initial.sum()
final=MC_Step(initial,10)
plt.hist(final,bins=10)
```
